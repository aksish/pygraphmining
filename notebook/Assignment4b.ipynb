```python
data1 = np.loadtxt('dataset/Dataset2Edges.txt')
```

```python
import networkx as nx
```

```python
graph = nx.from_edgelist(data1)
```

```python
laplacian1 = nx.laplacian_matrix(graph)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw_networkx(graph,with_labels=False,node_size=10,width=.3)
plt.show()
```

```python
eig,eiv = sp.linalg.eigh(laplacian1.todense())
```

```python
sorted_eig_arg = np.argsort(eig)
```

```python
np.dot(eiv[:,1],eiv[:sorted_eig_arg,2])
```

```python
sorted_eig = eig[sorted_eig_arg[::1]]
sorted_eiv = eiv[sorted_eig_arg[::1]]
```

```python
first_vec = sorted_eiv[:,0]
second_vec = sorted_eiv[:,1]
third_vec = sorted_eiv[:,2]
fourth_vec = sorted_eiv[:,3]
```

```python
x = np.column_stack((second_vec,third_vec))
kmeans = KMeans(n_clusters=4).fit(x)
labels = kmeans.labels_
```
